## Who was Kardashev
Nikolai Kardashev, born in 1932, was a **Soviet astronomer** who graduated from Moscow State University in 1955. After working at the Sternberg Astronomical Institute under the direction of Iosif Shklovsky, he obtained his doctorate in 1962 and then joined the Space Research Institute of the **Russian Academy of Sciences in Moscow**. In 1963 he led the first search for extraterrestrial signals in the Soviet Union, becoming one of the pioneers of the worldwide SETI program.
